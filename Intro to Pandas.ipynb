### Import Modules

```python
import pandas as pd
import numpy as np
```

## Create dataframe (that we will be importing)

```python
raw_data = {'first_name': ['Tendai', 'Kuda', 'Anna', 'Jessica', 'Gabriel'], 
        'last_name': ['Mare', 'Sizi', "Muku", 'Upani', 'Nhinda'], 
        'age': [29, 30, 25, 25, 30], 
        'TestScore': [40, 70, 80, 100, 0]
       }
df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'TestScore'])
df
```

```python
df.to_csv('example.csv')
```

## Load a csv

```python
df = pd.read_csv('example.csv')
df
```

## if you dont specify headers, first line is taken as the default header

```python
df = pd.read_csv('example.csv', names=['UID', 'First Name', 'Last Name', 'Age', 'Test Score'])
df
```

## Load a csv with setting the index column to UID

```python
df = pd.read_csv('example.csv', index_col='UID', names=['UID', 'First Name', 'Last Name', 'Age', 'Test Score'])
df
```

## Load a csv while setting the index columns to First Name and Last Name

```python
df = pd.read_csv('example.csv', index_col=['First Name', 'Last Name'], names=['UID', 'First Name', 'Last Name', 'Age', 'Test Score'])
df
```

## Drop a row by row number

```python
df.drop(df.index[0])
```

## Drop a variable (column)

Note: axis=1 denotes that we are referring to a column, not a row

```python
df.drop('Test Score', axis=1)
```

## fill in NAN fields

```python
df.fillna(0)
```
